### Checking Missing Values

```python
heart.isnull().sum()
```

```python
heart.isnull().mean()*100
```

Column	| Missing | %
|---|---|--|
PriorityArea2	| 42,640 | 100%
PriorityArea4	| 42,640 | 100%

No information → dropping them

Column	| Missing | %
|---|---|---|
PriorityArea1	| 33,280 | 78%
PriorityArea3	| 28,080 | 65%

Since they don’t affect heart disease measurement, dropping them 

```python
heart.drop(['PriorityArea1', 'PriorityArea2', 'PriorityArea3', 'PriorityArea4'], axis=1, inplace=True)
```

Column	| Missing	| %
|---|---|---|
Data_Value_Footnote_Symbol	| 42,111| 98.8%
Data_Value_Footnote	| 42,111 | 98.8%

These are just notes/explanations → not useful for EDA, dropping them

```python
heart.drop(['Data_Value_Footnote', 'Data_Value_Footnote_Symbol'], axis=1, inplace=True)
```

Column	| Missing	| %
|---|---|---|
Data_Value	| 529 | 1.2%
LowConfidenceLimit	| 529 | 1.2%
HighConfidenceLimit	| 529 | 1.2%
GeoLocation	| 820 | 1.9%

- These are important variables.

<li>Data_Value - the main measurement, cannot guess it, so removing missing rows
<li>ConfidenceLimits(Low, High) - since they are correspond to Data_value missing rows, automatically removed from Data_value
<li>Geolocation - Keep the rows, Ignore GeoLocation, the missing values are completely based for US, and Because the US is an aggregated national-level record with no single geographic coordinate. This is structural missingness, not a data error.

```python
heart = heart.dropna(subset = ['Data_Value'])
```

```python
heart[heart['GeoLocation'].isnull()]['LocationAbbr'].unique()
```

```python
heart.info()
```

### Check Duplicates

```python
heart.duplicated().sum()
```

- No Duplicates found in the dataset

Keeping only meaningful columns which provide analytical insights into heart disease trends

```python
keep_cols = [
    'Year',
    'LocationDesc',
    'LocationAbbr',
    'Topic',
    'Indicator',
    'Break_Out_Category',
    'Break_Out',
    'Data_Value',
    'Data_Value_Unit',
    'Data_Value_Type',
    'LowConfidenceLimit',
    'HighConfidenceLimit'
]

heart = heart[keep_cols]
heart.info()
```

### Outlier Detection

```python
sns.boxplot(x=heart['Data_Value'])
```

```python
outliers = heart[heart['Data_Value'] > heart['Data_Value'].quantile(0.99)]
outliers[['Topic','Break_Out_Category','Break_Out','Indicator','Data_Value']]
```

```python
heart.groupby(['Break_Out_Category','Break_Out'])['Data_Value'].median()
```

After careful analysis:
- The outliers were retained in the dataset.

Reason:
- They represent real high-risk populations
- Removing them would remove important epidemiological insights
- Dataset contains aggregated public health statistics
- This decision preserves the integrity of the analysis.

## Exploratory Data Analysis (EDA)

### Univariate Analysis

#### - Data_Value Distribution

```python
plt.figure()
plt.hist(heart['Data_Value'], bins=30)
plt.title("Distribution of Heart Disease Rates")
plt.xlabel("Data_Value")
plt.ylabel("Frequency")
plt.show()

plt.figure()
plt.boxplot(heart['Data_Value'])
plt.title("Boxplot of Heart Disease Rates")
plt.show()
```

Observations:
- The distribution is right-skewed.
- Majority of values lie between 0–30.
- Extreme high values exist (up to ~95).
- These extreme values correspond to hospitalization rates in elderly populations.

Interpretation:
- The skewness indicates that while most groups have moderate rates, a few demographic groups have very high disease burden.

#### - Topic Distribution

```python
heart['Topic'].value_counts().plot(kind='bar')
plt.title("Frequency of Disease Topics")
plt.show()
```

Observations:

Major topics include:
- Major Cardiovascular Disease
- Heart Failure
- Diseases of the Heart
- Coronary Heart Disease
- Stroke

Insight:

- The dataset heavily focuses on cardiovascular conditions, with heart failure and major cardiovascular disease appearing frequently.

#### - Break_Out_Category

```python
heart['Break_Out_Category'].value_counts().plot(kind='pie', autopct = '%1.1f%%')
plt.title("Demographic Category Distribution")
plt.show()
```

Categories include:

- Age
- Gender
- Race
- Overall

Insight:

- The dataset is structured to compare disease burden across demographic dimensions.

#### - Year Distribution

```python
heart['Year'].value_counts().sort_index().plot(kind='bar')
plt.title("Data Records per Year")
plt.show()
```

Insight:

- Data spans from 2004–2013, allowing time-series analysis.

### Bivariate Analysis

#### - Year vs Heart Disease Trend

```python
plt.figure(figsize = (5, 5))
heart.groupby('Year')['Data_Value'].mean().plot(marker='v')
plt.title("Average Heart Disease Rate Over Years")
plt.show()
```

Observations:

- Highest rate around 2004 (~17)
- Gradual decline until 2013 (~13.5)

Insight:

There is a clear decreasing trend in heart disease rates over time, possibly due to:
- Improved healthcare
- Awareness programs
- Preventive treatments

#### - Gender vs Heart Disease

```python
gender = heart[heart['Break_Out_Category']=='Gender']
gender.groupby('Break_Out')['Data_Value'].mean().plot(kind='bar')
plt.title("Heart Disease Rate by Gender")
plt.show()
```

Results:

- Male → 16.90
- Female → 13.61

Insight:

Males have significantly higher heart disease rates.

Possible reasons:
- Higher lifestyle risk exposure
- Biological risk factors

#### - Age vs Heart Disease

```python
age = heart[heart['Break_Out_Category']=='Age']
age.groupby('Break_Out')['Data_Value'].mean().plot(kind='bar')
plt.title("Heart Disease Rate by Age Group")
plt.show()
```

Results:

- 75+ → 10.26
- 65+ → 9.78

Insight:

- Risk increases with age, consistent with medical evidence.

#### - Race vs Heart Disease

```python
race = heart[heart['Break_Out_Category']=='Race']
race.groupby('Break_Out')['Data_Value'].mean().plot(kind='barh')
plt.title("Heart Disease Rate by Race")
plt.show()
```

Results:

- Non-Hispanic Black → 18.71 (Highest)
- Hispanic → 16.01
- Non-Hispanic White → 14.61
- Other → 14.58

Insight:

- Significant racial disparities exist in cardiovascular disease burden.
- This is a major public health observation

#### - Topic vs Heart Disease Rate

```python
heart.groupby('Topic')['Data_Value'].mean().sort_values(ascending=False).plot(kind='bar')
plt.title("Average Rate by Disease Type")
plt.show()
```

Results:

- Major Cardiovascular Disease → 23.84
- Heart Failure → 19.43
- Diseases of the Heart → 17.76
- Coronary Heart Disease → 6.54
- Stroke → 3.91

Insight:

- Heart failure and major cardiovascular disease are the most severe contributors.

### Multivariate Analysis

#### - Gender Trend Over Time

```python
gender_year = heart[heart['Break_Out_Category']=='Gender'].groupby(['Year','Break_Out'])['Data_Value'].mean().unstack()

gender_year.plot()
plt.title("Gender-wise Trend Over Time")
plt.show()
```

Observations:

- Both male and female rates decline over time.
- Males consistently remain higher.

Insight:

- Although overall rates decrease, gender disparity persists.

#### - Topic vs Race Heatmap

```python
pivot = heart[heart['Break_Out_Category']=='Race'].pivot_table(values='Data_Value', 
                                                               index='Topic', 
                                                               columns='Break_Out', 
                                                               aggfunc='mean')

sns.heatmap(pivot, annot=True)
plt.title("Disease Type vs Race")
plt.show()
```

Insight:

- Certain races show higher burden for specific diseases, especially heart failure.
- This highlights inequality patterns.

#### - Age + Topic Interaction

```python
pivot_age = heart[heart['Break_Out_Category']=='Age'].pivot_table(values='Data_Value', 
                                                                  index='Topic', 
                                                                  columns='Break_Out', 
                                                                  aggfunc='mean')

sns.heatmap(pivot_age, annot= True)
plt.title("Disease Type vs Age Group")
plt.show()
```

Insight:

- Heart failure hospitalization dramatically increases for 75+ age group.

#### - Year + Topic Trend

```python
topic_year = heart.groupby(['Year','Topic'])['Data_Value'].mean().unstack()
topic_year.plot()
plt.title("Disease Trend Over Time")
plt.legend(loc = 'upper left', bbox_to_anchor=(1, 1))
plt.show()
```

Insight:

- Some diseases decline faster than others.

## Pivot Table

### - Year vs Heart Disease Rate

```python
pivot_year = pd.pivot_table(
    heart,
    values='Data_Value',
    index='Year',
    aggfunc='mean'
)
pivot_year
```

```python
pivot_year.plot(marker='o')
plt.title("Average Heart Disease Rate Over Years")
plt.ylabel("Average Data_Value")
plt.show()
```

#### Insight

- Heart disease rates decreased from 2004 to 2013.
- Early years showed higher average values (~17).
- Later years dropped to ~13.5.

#### Interpretation

This decline suggests:

- Improved medical treatment
- Better awareness campaigns
- Lifestyle improvements
- Early diagnosis programs

Indicates progress in cardiovascular healthcare.

### - Gender vs Heart Disease Type

```python
gender_df = heart[heart['Break_Out_Category']=='Gender']

pivot_gender = pd.pivot_table(
    gender_df,
    values='Data_Value',
    index='Break_Out',
    aggfunc='mean'
)
pivot_gender
```

```python
pivot_gender.plot(kind='bar')
plt.title("Heart Disease Rate by Gender")
plt.ylabel("Average Data_Value")
plt.xticks(rotation=0)
plt.show()
```

#### Insight

- Males show higher rates across most disease types.
- Heart failure and major cardiovascular disease show biggest gender gap.

#### Interpretation

Possible causes:

- Higher smoking/alcohol history
- Occupational stress
- Biological risk factors

Gender-based preventive programs are needed.

### - Age Group vs Disease Type

```python
age_df = heart[heart['Break_Out_Category']=='Age']

pivot_age = pd.pivot_table(
    age_df,
    values='Data_Value',
    index='Break_Out',
    aggfunc='mean'
)

pivot_age
```

```python
pivot_age.plot(kind='bar')
plt.title("Heart Disease Rate by Age Group")
plt.ylabel("Average Data_Value")
plt.xticks(rotation=0)
plt.show()
```

#### Insight

- Age 75+ shows highest values across almost all diseases.
- Heart failure increases sharply with age.

#### Interpretation

- Cardiovascular system weakens with age.

Elderly population needs regular heart screening.

### - Race vs Disease Type

```python
race_df = heart[heart['Break_Out_Category']=='Race']

pivot_race = pd.pivot_table(
    race_df,
    values='Data_Value',
    index='Break_Out',
    aggfunc='mean'
)

pivot_race
```

```python
pivot_race.plot(kind='bar', color = ['coral'])
plt.title("Heart Disease Rate by Race")
plt.ylabel("Average Data_Value")
plt.xticks(rotation=45)
plt.show()
```

#### Insight

- Non-Hispanic Black population has highest rates in most disease categories.
- Hispanic population also shows high hospitalization rates.
- Non-Hispanic White population has relatively lower average rates.

#### Interpretation

- Possible reasons:
- Socioeconomic factors
- Healthcare access inequality
- Lifestyle differences
- Genetic predisposition

Shows need for targeted public health policies.

### - Indicator vs Demographics

```python
pivot_topic = pd.pivot_table(
    heart,
    values='Data_Value',
    index='Topic',
    aggfunc='mean'
).sort_values(by='Data_Value', ascending=False)

pivot_topic
```

```python
pivot_topic.plot(kind='bar')
plt.title("Average Rate by Disease Type")
plt.ylabel("Average Data_Value")
plt.show()
```

#### Insight

- Hospitalization indicators show very high values for elderly.
- Mortality indicators are lower but consistent.
- Prevalence indicators vary by race and gender.

#### Interpretation

- Different indicators measure different aspects of heart disease.

Hospitalization rates highlight severe cases.

### - Topic vs Year

### - Location vs Disease Rate

```python
pivot_location = pd.pivot_table(
    heart,
    values='Data_Value',
    index='LocationDesc',
    aggfunc='mean'
).sort_values(by='Data_Value')

pivot_location
```
